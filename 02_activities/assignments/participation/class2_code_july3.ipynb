```python
# Importing the necesssary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613) # Adding this line for reproducibility
x = np.arange(50) # Creating an array of integers starting from 0 up to (but not including) 50
y = np.random.randint(0,100,50) # Generating 50 random integers between 0 (inclusive) and 100 (exclusive)

fig, ax = plt.subplots(figsize=(5,3)) # Creating a figure with axes (plots) and the size of the figure is 5 inches wide and 3 inches tall
#ax.scatter(x,y) # Creating a scatterplot of x and y 
#ax.bar(x,y) # Creating a vertical bar chart of x and y
#ax.barh(x,y) # Creating a horizontal bar chart of x and y 
#ax.plot(x,y) # Creating a line plot of x an y, connecting each point with lines
ax.hist(y) # Creating a histogram of y 
```

```python
font1 = {'family':'sans-serif', 'color':'blue', 'size':20} # Creating a font dictionary with style, color and size, named font1 
font2 = {'family':'monospace', 'color':'green', 'size':14} # Creating a font dictionary with style, color and size, named font2 

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left') # Setting a title for the figure with font1 and changing its location as left
ax.set_ylabel('Total growth', fontdict = font2) # Setting a name for y axis with font2 style, color and size
ax.set_xlabel('Years since start', fontdict = font2) # Setting a name for x axis with font2 style, color and size
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y, 
           marker = '*',
           color = 'indigo') # Changing the marker of the scatterplot as * and the color as indigo
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y, 
        marker = '*',
        color = 'indigo', 
        linestyle = '--', 
        linewidth = 2) # Changing the marker of the plot as *, color as indigo, linestyle as -- and linewidth as 2
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y, 
        marker = '*',
        color = '#7425B9', 
        linestyle = '--', 
        linewidth = 2,
        markersize = 12,
        markeredgecolor = '#FA9359',
        markerfacecolor = '#000000') # Changing the colors using hex codes, and also changing other components of style
ax.grid(axis = 'y', 
        color = 'blue', 
        linestyle = '--',
        linewidth = 2) # Adding grids on y axis and changing the grid style with blue as color, -- as linestyle and 2 as linewidth 
```

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
 
# --- FORMAT 1
 
# Creating our x and y axis
x=range(1,6)
y=[ [1,4,6,8,9], [2,2,7,10,12], [2,8,5,10,6] ]
 
# Creating a basic stacked area chart
plt.stackplot(x,y, labels=['A','B','C'])
plt.legend(loc='upper left')
plt.show()
 
# --- FORMAT 2 with different y's
x=range(1,6)
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]
 
# Creating a basic stacked area chart
plt.stackplot(x,y1, y2, y3, labels=['A','B','C'])
plt.legend(loc='upper left')
plt.show()
```
